# ML concepts overview

<br/>1. ML - automated analytical models
<br/> 2. Neural n/w - a type of ML architecture modeled after biological neurons
<br/>3. DL - a neural network with more than 1 hidden layer\n
# <br/>--------------supervised -----learning----------------------------------
<br/>trained on labelled input.
<br/>+ve /-ve.


<br/> data needs to be split in  3 sets:
<br/> 1. Training data - train model parameters.
<br/> 2. Validation data - determine what model hyperparameters to adjust.
<br/> 3. Test data - get some final performance metric.
# <br/>----------------- Overfitting-------------------------------------------------
<br/>- Too much to the noise from data
<br/>- The overfitting will usually results in low error on training set, but high error on test/validation sets.
# <br/> ----------------Underfitting------------------------------------------------
<br/>- models won't be able to capture the trend data 
<br/> - does not fit the given data well enough.
<br/> - it is oftenly result of simple model.
# <br/> ------------- how to measure under/overfitting?---------------------------------------
<br/> -  good model -> decrease err wrt train times-epocs.
<br/> - plot the er v/s epocs chart 
# <br/> ------------- Evaluating Performance - classification error metrics-----------------------
<br/> - Accuracy, precesion,recall and F1 score.
<br/> - binary classification
<br/> - Accuracy - num of correct predictions/num of total predictions.
<br/> --- Accuracy are used for well balanced classes.
<br/> - Recall  -  TP/(TP+FN) -> says overall data points
<br/> - precision - TP/(TP+FP) -> says which are actually relevent.
<br/> - F1 score is combination of recall and precision --> harmonic mean of precision and recall.
<br/> -------------F1=2*((precision*recall)/(precision+recall))
<br/> --it punishes extream differences between recall and precision.
### <br/>  confusion matrix:

<br/> -we need to concentrate on false positives , not on classifying false negatives

# <br/> ------------- Evaluating Performance - regression error metrics-----------------------
<br/> - recall. accuracy are not useful for regression, bcoz regressor gives continuous values.
<br/> - Some of the most common evaluation metrics for regression models:
<br/> --------Mean absolute error.
<br/> --------Mean squared error.
<br/> --------Root mean square error.

## Mean absolute error (MAE)
<br/> - 
<br/> - MAE won't punish large errors, so when we take the squared values of(yi-y^) then outlayers and large errors are noted 
##  Mean squar error (MSE)
<br/> - we take positive error , however we square units as well, so to get smaller values we take root on it
## Root Mean square error(RMSE)
<br/> - take root after calculating mean square error.
<br/> - Compare your value with the average value of the label to get intution of overall performance.

# --------------unsupervised -----learning----------------------------------
<br/> - tasks which fall under unsupervised learning :
<br/> -------Clustering(based on similarity)
<br/> -------Anomaly detection(outliers - anamoly)
<br/> -------Dimensionality reduction(reduce number of features which are imoratant in a dataset)

## Regression code

```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
         batch_size=128,
         epochs=400)
```

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,8))
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```
